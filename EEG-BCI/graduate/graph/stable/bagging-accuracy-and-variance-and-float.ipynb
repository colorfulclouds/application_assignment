## sbp

```python
#accuracy
dt = np.array([[0.7185676392572944 ],
               [0.7121352785145889 ],
               [0.7423740053050397 ],
               [0.721153846153846 ]])
                
svm = np.array([[0.5807029177718833 ],
               [0.5754641909814323 ],
               [0.5985411140583554 ],
               [0.5787135278514589 ]])
                
nb = np.array([[0.1130636604774536],
               [0.20384615384615384 ],
               [0.1142572944297082 ],
               [0.11246684350132626 ]])

knn = np.array([[0.6428381962864721 ],
               [0.6438992042440319 ],
               [0.6626657824933687 ],
               [0.6470159151193633 ]])

dt = dt * 100
svm = svm * 100
nb = nb * 100
knn = knn * 100

freq = [6 , 7.5 , 8.5 , 10]
plt.figure(figsize=(2.4,2.0) , dpi=100) #2.2,2.0) , dpi=115

plt.plot(freq , dt , marker='o' , label='决策树')
plt.plot(freq , svm , marker='s' , label='SVM')
plt.plot(freq , nb , marker='p' , label='朴素贝叶斯')
plt.plot(freq , knn , marker='*' , label='kNN')

#plt.legend(loc=3)
plt.ylabel('准确率(%)',fontsize=10.5)
plt.xlabel('刺激频率(Hz)',fontsize=10.5)

plt.xticks(freq)

plt.title('SBP',fontsize=10.5)

plt.grid() #去掉网格线
plt.show()
plt.close()
```
