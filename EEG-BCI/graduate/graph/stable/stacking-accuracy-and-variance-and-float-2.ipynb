```python
'''
base
'''
dt = np.array([[0.00020845499511007578],
               [0.0003186190010483434],
               [0.0002749966579656507],
               [0.0004081503422946757]])
                
svm =np.array([[0.0001372476764066449],
               [0.00017664850241681882],
               [0.0002932679115451453],
               [0.0002698736711016049]])
                
nb = np.array([[0.0010123681655397562],
               [0.0003681514680325619],
               [0.0009715249878631384],
               [0.0011016884309324629]])

knn =np.array([[0.00023367855961837518],
               [0.0002640690851268917],
               [0.0003254789662911862],
               [0.00016532516235251042]])

tt = np.array([dt , svm , nb , knn]).squeeze()
base_variance_mean = np.mean(tt , axis=0)

'''
bagging
'''

dt_b = np.array([[0.0003890127982325917],
                 [0.00018618209513892307],
                 [0.0005433224394739988],
                 [0.00020964669420033924]])
                
svm_b =np.array([[0.0005224611092739692],
                 [0.00029654398468996484],
                 [0.0003046704050545636],
                 [0.00038069412294464902]])
                
nb_b = np.array([[0.0002318096588310619],
                 [0.0005783654285895208],
                 [0.00023539794834270272],
                 [0.0002252364049560608]])

knn_b =np.array([[0.0002332564079111231],
                 [0.000234382145797128],
                 [0.00025861629224155456],
                 [0.0003256328757677887]])

tt = np.array([dt_b , svm_b , nb_b , knn_b]).squeeze()
bagging_variance_mean = np.mean(tt , axis=0)

'''
stacking
'''

sbp_s = np.array([[0.00032967849629561894],
                  [0.00032890455149899055],
                  [0.00020605840468869836],
                  [0.00010696268882494085]])

'''
===========
'''

freq = [6 , 7.5 , 8.5 , 10]
plt.figure(figsize=(2.4,2.0) , dpi=100) #2.2,2.0) , dpi=115

plt.plot(freq , base_variance_mean , marker='o' , label='Base')
plt.plot(freq , bagging_variance_mean , marker='s' , label='Bagging')
plt.plot(freq , sbp_s , marker='*' , label='Stacking')

#plt.legend(loc=3)
plt.ylabel('方差',fontsize=10.5)
plt.xlabel('刺激频率(Hz)',fontsize=10.5)

plt.xticks(freq)

plt.title('SBP',fontsize=10.5)

plt.grid() #去掉网格线
plt.show()
plt.close()
```
