```python
#accuracy
dt = np.array([[0.6423076923076925],
               [0.6294429708222813],
               [0.6610079575596817],
               [0.653315649867374]])
                
svm =np.array([[0.6545092838196285],
               [0.6629310344827586],
               [0.6894562334217506],
               [0.6677055702917772]])
                
nb = np.array([[0.1300397877984085],
               [0.13395225464190982],
               [0.1622679045092838],
               [0.11226790450928381]])

knn =np.array([[0.620026525198939],
               [0.6403846153846154],
               [0.6705570291777189],
               [0.6484084880636605]])

dt = dt * 100
svm = svm * 100
nb = nb * 100
knn = knn * 100

freq = [6 , 7.5 , 8.5 , 10]

#=========
plt.figure(figsize=(2.4,2.0) , dpi=100) #2.2,2.0) , dpi=115
#=========

plt.plot(freq , dt , marker='o' , label='决策树')
plt.plot(freq , svm , marker='s' , label='SVM')
plt.plot(freq , nb , marker='p' , label='朴素贝叶斯')
plt.plot(freq , knn , marker='*' , label=r'kNN')

#plt.legend(loc=3 , fontsize =10.5)

plt.ylabel('准确率(%)',fontsize=10.5)
plt.xlabel('刺激频率(Hz)',fontsize=10.5)#

plt.xticks(freq)

plt.title('SBP' , fontsize=10.5)


plt.grid() #去掉网格线
plt.savefig(fname='1.svg', dpi=100)
plt.show()
plt.close()
```
