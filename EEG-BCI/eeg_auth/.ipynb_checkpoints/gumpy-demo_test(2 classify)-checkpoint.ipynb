```python
class EEG(gp.data.Dataset):
    def __init__(self , base_dir='default' , identifier='default' , **kwargs):
        super(EEG, self).__init__(**kwargs)
        
        self.base_dir = 'no use'
        self.data_id = 'no use'
        self.data_dir = 'no use'
        self.data_type = 'EEG'
        self.data_name = 'EEG'
        
        
        # length of a trial (in seconds)
        self.trial_len = 3
        # motor imagery appears in interval (in seconds)
        self.mi_interval = [0, 2]
        # idle perior prior to start of signal (in seconds)
        self.trial_offset = 0
        # total length of a trial (in seconds)
        self.trial_total = self.trial_len
        # sampling frequency (in Hz)
        self.expected_freq_s = 128 #老设备 新设备为256
    
    def _sep(self , data , label):
        train_data = []
        train_labels = []

        size = self.expected_freq_s*self.trial_len
        data_length = data.shape[0]

        idx = 0
        
        while idx<data_length-size:
            train_data.append(data[idx : idx+size , :])
            train_labels.append(label)

            #no overlap
            idx = idx + size
        
        return np.array(train_data) , np.array(train_labels)
    
    def _load_data(self , filename):
        data = pd.read_csv(filename)

        #删除前3秒和后2秒数据
        data = data.iloc[ : data.shape[0] - 2*self.expected_freq_s] #后2秒
        data = data.iloc[3*self.expected_freq_s : ] #前3秒

        return data[['AF3' , 'T7' , 'Pz' , 'T8' , 'AF4']].values
    
    def _combine(self):
        data1 = self._load_data('data/train_1/fei_pink_1.csv')
        data2 = self._load_data('data/train_1/sen_pink_1.csv')

        #数据分段
        train_data1 , train_labels_1 = self._sep(data1 , label=0)
        train_data2 , train_labels_2 = self._sep(data2 , label=1)

        #产生索引并置乱
        idx_train_data = list(range(len(train_data1) + len(train_data2)))
        np.random.shuffle(idx_train_data)

        #合并数据
        train_data = np.concatenate((train_data1 , train_data2))
        train_labels = np.concatenate((train_labels_1 , train_labels_2))

        #将训练数据打乱
        train_data = train_data[idx_train_data]
        train_labels = train_labels[idx_train_data] ##self.labels

        #合并为sequence
        train_data_sequence = np.reshape(train_data , newshape=(-1 , origin_channel)) #length*channel

        train_data_count = train_data.shape[0]

        #trial索引
        trial_idx = np.array([i*(sample*trial) for i in range(train_data_count)]) ##self.trial
        
        #构建子类成员
        self.raw_data = train_data_sequence
        self.trials = trial_idx
        self.labels = train_labels
        
        
    def load(self, **kwargs):
        self._combine()
        
        self.sampling_freq = self.expected_freq_s
        
        
        
    '''
    def _load(self):
        trials = []
        labels = []
        
        filenames = self.filenames
        
        data = self._load_data(filenames[0])
        
        label_idx = 0
        trial , label = self._sep(data , label_idx)
        label_idx = label_idx + 1
        
        trials.append(trial)
        labels.append(label)

        filenames.remove(filenames[0])

        for filename in filenames:
            data = self._load_data(filename)
            trial , label = self._sep(data , label_idx)
            label_idx = label_idx + 1
            
            trials.append(trial)
            labels.append(label)

        self.trials = np.array(trials)
        self.labels = np.array(labels)
    '''
    
```

# Alpha and Beta sub-bands

```python
def alpha_subBP_features(data):
    alpha1 = gp.signal.butter_bandpass(data , 8.5,11.5 , order=6)
    alpha2 = gp.signal.butter_bandpass(data , 9.0,12.5 , order=6)    
    alpha3 = gp.signal.butter_bandpass(data , 9.5,11.5 , order=6)    
    alpha4 = gp.signal.butter_bandpass(data , 8.0,10.5 , order=6)   
    
    return [alpha1 , alpha2 , alpha3 , alpha4]

alpha_bands = np.array(alpha_subBP_features(dataset))
```

```python
def beta_subBP_features(data):
    beta1 = gp.signal.butter_bandpass(data , 14.0,30.0,order=6)
    beta2 = gp.signal.butter_bandpass(data , 16.0,17.0,order=6)    
    beta3 = gp.signal.butter_bandpass(data , 17.0,18.0,order=6)    
    beta4 = gp.signal.butter_bandpass(data , 18.0,19.0,order=6)    
    
    return [beta1 , beta2 , beta3 , beta4]

beta_bands = np.array(beta_subBP_features(dataset))
```

## Method 1: logarithmic sub-band power

```python
def powermean(data , trial , fs , w):
    #官方demo跳4秒 前4秒为准备阶段
    return np.power(data[trial + w[0] : trial+w[1] , 0] , 2).mean(), \
            np.power(data[trial + w[0] : trial+w[1] , 1] , 2).mean(), \
            np.power(data[trial + w[0] : trial+w[1] , 2] , 2).mean(), \
            np.power(data[trial + w[0] : trial+w[1] , 3] , 2).mean(), \
            np.power(data[trial + w[0] : trial+w[1] , 4] , 2).mean()
```

```python
def log_subBP_feature_extraction(alpha , beta , trials , fs , w):
    #特征数的2倍
    n_features = 25
    X = np.zeros((len(trials) , n_features))
    
    for t , trial in enumerate(trials):
        # trial是一个scalar
        # t也是一个scalar
        power_AF3_1, power_T7_1, power_Pz_1 , power_T8_1 , power_AF4_1 = powermean(alpha[0], trial, fs, w)
        power_AF3_2, power_T7_2, power_Pz_2 , power_T8_2 , power_AF4_2 = powermean(alpha[1], trial, fs, w)
        power_AF3_3, power_T7_3, power_Pz_3 , power_T8_3 , power_AF4_3 = powermean(alpha[2], trial, fs, w)
        power_AF3_4, power_T7_4, power_Pz_4 , power_T8_4 , power_AF4_4 = powermean(alpha[3], trial, fs, w)
        
        power_AF3_1_b, power_T7_1_b, power_Pz_1_b , power_T8_1_b , power_AF4_1_b = powermean(beta[0], trial, fs, w)
        
        X[t, :] = np.array(
            [np.log(power_AF3_1) , np.log(power_T7_1) , np.log(power_Pz_1) , np.log(power_T8_1) , np.log(power_AF4_1),
             np.log(power_AF3_2) , np.log(power_T7_2) , np.log(power_Pz_2) , np.log(power_T8_2) , np.log(power_AF4_2),
             np.log(power_AF3_3) , np.log(power_T7_3) , np.log(power_Pz_3) , np.log(power_T8_3) , np.log(power_AF4_3),
             np.log(power_AF3_4) , np.log(power_T7_4) , np.log(power_Pz_4) , np.log(power_T8_4) , np.log(power_AF4_4),
             np.log(power_AF3_1_b), np.log(power_T7_1_b), np.log(power_Pz_1_b) , np.log(power_T8_1_b) , np.log(power_AF4_1_b)])

    return X
```

```python
#采样率拆分成两部分

w1 = [0,384//2]
w2 = [384//2,384]

features1 = log_subBP_feature_extraction(alpha_bands , beta_bands , dataset.trials , dataset.sampling_freq , w1)
features2 = log_subBP_feature_extraction(alpha_bands , beta_bands , dataset.trials , dataset.sampling_freq , w2)
```

## Method 2: DWT

```python
def dwt_features(data , trials , level , sampling_freq , w , n , wavelet):
    n_features = 15
    
    X = np.zeros((len(trials) , n_features))
    
    for t , trial in enumerate(trials):
        signals = data[trial + w[0] : trial + w[1]]
        
        coeffs_AF3 = pywt.wavedec(data = signals[:,0], wavelet=wavelet, level=level)
        coeffs_T7 = pywt.wavedec(data = signals[:,1], wavelet=wavelet, level=level)
        coeffs_Pz = pywt.wavedec(data = signals[:,2], wavelet=wavelet, level=level)
        coeffs_T8 = pywt.wavedec(data = signals[:,3], wavelet=wavelet, level=level)
        coeffs_AF4 = pywt.wavedec(data = signals[:,4], wavelet=wavelet, level=level)

        X[t, :] = np.array([
            np.std(coeffs_AF3[n]), np.mean(coeffs_AF3[n]**2),  
            np.std(coeffs_T7[n]), np.mean(coeffs_T7[n]**2),
            np.std(coeffs_Pz[n]), np.mean(coeffs_Pz[n]**2), 
            np.std(coeffs_T8[n]), np.mean(coeffs_T8[n]**2),
            np.std(coeffs_AF4[n]), np.mean(coeffs_AF4[n]**2), 
            np.mean(coeffs_AF3[n]),
            np.mean(coeffs_T7[n]), 
            np.mean(coeffs_Pz[n]),
            np.mean(coeffs_T8[n]),
            np.mean(coeffs_AF4[n])])
        
        return X
```
