```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

# c分类法 m匹配法
# n不特征选择 y特征选择

```python

c_n_pre_10 = np.array([0.6498 , 0.8688 , 0.6635 , 0.8345])
c_y_pre_15 = np.array([0.5470 , 0.7465 , 0.5543 , 0.6928])
m_n_pre_20 = np.array([0.2462 , 0.3077 , 0.3442 , 0.2958])
m_y_pre_25 = np.array([0.3570 , 0.2251 , 0.2433 , 0.2093])

freq = [10,15,20,25]
```

```python
plt.plot(freq , c_n_pre_10 , marker='o' , label='c_n')
plt.plot(freq , c_y_pre_15 , marker='s' , label='c_y')
plt.plot(freq , m_n_pre_20 , marker='p' , label='m_n')
plt.plot(freq , m_y_pre_25 , marker='*' , label='m_y')

plt.legend(loc=3)
plt.ylabel('Precision')
plt.xlabel('Freq(Hz)')

plt.xticks(freq)

plt.grid() #去掉网格线
```
